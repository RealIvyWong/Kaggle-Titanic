## 2 特征工程
### 2.1 构造新特征
参考相关分析，考虑构建以下新特征：
* NameLen => 姓名长度
* HasCabin => 是否有客舱信息
* FamilySize => 家庭大小 （FamilySize = SibSp + Parch + 1）
* IsAlone => 是否独行（家庭大小是否为1）
* Title => 姓名的头衔
* CabinNum => 客舱号
* CateAge => 年龄分层

```python
# NameLen 姓名长度
train['NameLen'] = train['Name'].apply(len)

# HasCabin 是否有客舱信息
train['HasCabin'] = train["Cabin"].apply(lambda x: 0 if type(x) == float else 1)

# FamilySize 家庭大小
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1

# IsAlone 是否独行
train['IsAlone'] = train["FamilySize"].apply(lambda x: 1 if x==1 else 0)

# Title 姓名的头衔
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    # If the title exists, extract and return it.
    if title_search:
        return title_search.group(1)
    return ""
train['Title'] = train['Name'].apply(get_title)

# CabinNum 客舱号
train['CabinNum']=train['Cabin'].apply(lambda x: np.nan if type(x)==float else x[0])

# 删除其他特征
train.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)

train.head()
```

#### 对于缺失值的处理
缺失值的几种常见处理方式：
* 如果缺值的样本占总数比例极高，我们可能就直接舍弃了，作为特征加入的话，可能反倒带入 noise，影响最后的结果
* 如果缺值的样本适中，而该属性非连续值特征属性（比如说类目属性），那就把 NaN 作为一个新类别，加到类别特征中
* 如果缺值的样本适中，而该属性为连续值特征属性，有时候我们会考虑给定一个 step，然后把它离散化，之后把 NaN 作为一个 type 加到属性类目中
* 有些情况下，缺失的值个数并不是特别多，那我们也可以试着根据已有的值，拟合一下数据，补充上

先查看一下都有哪些字段有缺失值。

```python
train.isnull().sum().to_frame().T
```

目前存在缺失值的有**年龄(Age)**，**登船港口(Embarked)**和**客舱号(CabinNum)**。
* 对于登船港口(Embarked)，我们把缺失的 2 条数据填充为出现得最多的 S 港口
* 年龄(Age)缺失的适中，而且是连续变量，希望通过数据之间的联系进行拟合填充（当然，第三种方法也是可以尝试的）
* 客舱号(CabinNum)几乎缺失了 2/3，考虑要直接舍弃了

```python
# 填充登船港口(Embarked)
train['Embarked']=train['Embarked'].fillna('S')
train.isnull().sum().to_frame().T
```

在考虑填充年龄(Age)和处理客舱号(CabinNum)之前，先看看各个特征之间的相关关系。<br>
而发现这些特征中性别(Sex)、登船港口(Embarked)、头衔(Title)和船舱号(CabinNum)都还是字符串变量，把他们进行转换数字表示。
* 关于头衔(Title)，将图中出现较少的统一归为 Rare 组。

```python
train['Sex'] = train['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
train['Embarked'] = train['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
train['Title'] = train['Title'].replace(['Dr','Rev','Mlle','Major','Col','Don','Capt','Sir','Lady','Jonkheer','Ms','Mme','Countess'], 'Rare')
train['Title'] = train['Title'].map( {'Mr': 0, 'Miss': 1, 'Mrs': 2, 'Master': 3, 'Rare':4} ).astype(int)
train['CabinNum'] = train['CabinNum'].replace(['A','B','C','D','E','F','G','T'], [0,1,2,3,4,5,6,7])
train.head()
```

再看看，这些特征之间的相关关系。

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(train.astype(float).corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap='coolwarm', linecolor='white', annot=True)
plt.show()
```

可以发现年龄(Age)与其他特征都没有很强的线性相关关系，而客舱号(CabinNum)与客舱等级(Pclass)比较强的线性相关关系。<br>
所以我们尝试使用其他特征对年龄进行拟合。删除客舱号，因为缺失数据太多，而且有其他特征可以表达它的大部分信息，强行拟合倒是会带入噪音。

```python
# 删除客舱号(CabinNum)
train.drop("CabinNum",axis=1,inplace=True)
```

```python
# 使用回归决策树对年龄(Age)进行拟合

#将已经存在数值特征取出来
age_df=train.drop("Survived",axis=1)
#将乘客分成已知年龄和未知年龄两部分
known_age=age_df[age_df.Age.notnull()].as_matrix()
unknown_age=age_df[age_df.Age.isnull()].as_matrix()
#y为目标年龄
y=known_age[:,2]
#X为特征属性值
X=np.delete(known_age,2,axis=1)
# fit到DecisionTreeRegressor之中
dtr=DecisionTreeRegressor(max_depth=4)
dtr.fit(X,y)

#用得到的模型进行未知年龄结果预测
predictedAges=dtr.predict(np.delete(unknown_age,2,axis=1))
    
#用得到的结果填补原缺失数据
train.loc[(train.Age.isnull()),'Age']=predictedAges
```

```python
train.isnull().sum().to_frame().T
```

```python
train.head()
```

接下来，对特征中的多类别变量进行**因子化**。<br>
需要处理的有客舱等级(Pclass)，登船港口(Embarked)和头衔(Title)。

```python
dummies_Pclass=pd.get_dummies(train['Pclass'],prefix='Pclass')
dummies_Embarked=pd.get_dummies(train['Embarked'],prefix='Embarked')
dummies_Title=pd.get_dummies(train['Title'],prefix='Title')
train=pd.concat([train,dummies_Pclass,dummies_Embarked,dummies_Title],axis=1)
train.drop(['Pclass','Embarked','Title'],axis=1,inplace=True)
train.head()
```

对于数值型特征变化幅度较大的年龄(Age)，票价(Fare)和姓名长度(NameLen)还需要做一个标准化 scaling。

```python
scaler = preprocessing.StandardScaler()
age_scale_param = scaler.fit(train['Age'].values.reshape(-1,1))
train['Age_scaled'] = scaler.fit_transform(train['Age'].values.reshape(-1,1), age_scale_param)
fare_scale_param = scaler.fit(train['Fare'].values.reshape(-1,1))
train['Fare_scaled'] = scaler.fit_transform(train['Fare'].values.reshape(-1,1), fare_scale_param)
namelen_scale_param = scaler.fit(train['Fare'].values.reshape(-1,1))
train['NameLen_scaled'] = scaler.fit_transform(train['NameLen'].values.reshape(-1,1), namelen_scale_param)
train.drop(['Age','Fare','NameLen'],axis=1,inplace=True)
train.head()
```

## 3 建模

```python
%%time
# 逻辑回归建模
train_np=train.as_matrix()

y=train_np[:,0]
X=train_np[:,1:]

x_train, x_val, y_train, y_val = train_test_split(X, y, test_size = 0.22, random_state = 0)

clf=LogisticRegression(C=1.0,penalty='l1',tol=1e-6)
clf.fit(x_train, y_train)

y_pred = clf.predict(x_val)

# 准确度
acc_logreg = round(accuracy_score(y_pred, y_val) * 100, 2)
print("Accuracy:",acc_logreg)
```

```python
print(classification_report(y_val,y_pred))  ##(y_test,predictions)
```

```python
print(confusion_matrix(y_val, y_pred))  ##y_test, predictions
```

## 4 预测
首先对测试集进行和训练集同样的处理
* 添加新特征

```python
# NameLen 姓名长度
test['NameLen'] = test['Name'].apply(len)

# HasCabin 是否有客舱信息
test['HasCabin'] = test["Cabin"].apply(lambda x: 0 if type(x) == float else 1)

# FamilySize 家庭大小
test['FamilySize'] = test['SibSp'] + test['Parch'] + 1

# IsAlone 是否独行
test['IsAlone'] = test["FamilySize"].apply(lambda x: 1 if x==1 else 0)

# Title 姓名的头衔
test['Title'] = test['Name'].apply(get_title)

# 删除其他特征
test.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)
test.head()
```

* 处理缺失值

```python
test.isnull().sum().to_frame().T
```

发现有两个特征需要处理，年龄(Age)和票价(Fare)。<br>
对于年龄，使用训练集使用的回归决策树进行拟合。<br>
对于票价，训练集没有票价信息缺失，所以没有参考方法。但是测试集中只缺失 1 条数据，而从训练集的票价分布情况来看，票价都集中分布在一个峰，所以取众数是可以的。

先进行特征数字化。

```python
test['Sex'] = test['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
test['Embarked'] = test['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
def map_title(title):
    if title=='Mr':
        return 0
    elif title=='Miss':
        return 1
    elif title=='Mrs':
        return 2
    elif title=='Master':
        return 3
    else:
        return 4
test['Title'] = test['Title'].apply(map_title)
test.head()
```

```python
#预测年龄
unknown_age_test=test[test.Age.isnull()].as_matrix()
predictedAges_test=dtr.predict(np.delete(unknown_age_test,[1,3],axis=1))
    
#用得到的结果填补原缺失数据
test.loc[(test.Age.isnull()),'Age']=predictedAges_test
```

```python
# 填充票价
test.loc[(test.Fare.isnull()),'Fare']=float(test.Fare.mode())
```

```python
test.isnull().sum().to_frame().T
```

* 因子化

```python
dummies_Pclass_test=pd.get_dummies(test['Pclass'],prefix='Pclass')
dummies_Embarked_test=pd.get_dummies(test['Embarked'],prefix='Embarked')
dummies_Title_test=pd.get_dummies(test['Title'],prefix='Title')
test=pd.concat([test,dummies_Pclass_test,dummies_Embarked_test,dummies_Title_test],axis=1)
test.drop(['Pclass','Embarked','Title'],axis=1,inplace=True)
test.head()
```

* 标准化

```python
scaler = preprocessing.StandardScaler()
age_scale_param_test = scaler.fit(test['Age'].values.reshape(-1,1))
test['Age_scaled'] = scaler.fit_transform(test['Age'].values.reshape(-1,1), age_scale_param_test)
fare_scale_param_test = scaler.fit(test['Fare'].values.reshape(-1,1))
test['Fare_scaled'] = scaler.fit_transform(test['Fare'].values.reshape(-1,1), fare_scale_param_test)
namelen_scale_param_test = scaler.fit(test['Fare'].values.reshape(-1,1))
test['NameLen_scaled'] = scaler.fit_transform(test['NameLen'].values.reshape(-1,1), namelen_scale_param_test)
test.drop(['Age','Fare','NameLen'],axis=1,inplace=True)
test.head()
```

* 预测

```python
idss = test['PassengerId']
predictions = clf.predict(test.drop(['PassengerId'], axis=1)).astype(int)
output = pd.DataFrame({"PassengerId":idss,"Survived":predictions})
```
